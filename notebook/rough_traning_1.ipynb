```python
def assign_points(percentage):
    if percentage < 33:
        return 1
    elif 33 <= percentage < 50:
        return 2
    elif 50 <= percentage < 75:
        return 3
    elif 75 <= percentage < 100:
        return 4
    elif percentage == 100:
        return 5
```

```python
merged_DocumentID['Points'] = merged_DocumentID['Percentage'].apply(assign_points)
```

```python
new_df = new_df.merge(merged_DocumentID[['DocumentID', 'Points']], on='DocumentID', how='left')
```

```python
new_df['Points'].fillna(0, inplace=True)
new_df['Points'] = new_df['Points'].astype(int)
```

```python
keywords = [
    'financial', 'information', 'bank', 'article', 'date', 
    'securities', 'republic', 'paragraph', 'credit', 'data', 
    'their', 'risk', 'section', 'services', 'legal', 
    'accordance', 'reporting', 'all', 'state', 'foreign', 
    'person', 'market', '$', 'following', 'payment', 
    'investment', 'business', 'form', 'within', 'kazakhstan', 
    'management', 'provided', 'act', 'amount', 'requirements', 
    'account', 'exchange', 'service', 'public', 'electronic', 
    'national', 'case', 'been', 'into', 'tax', 'regulation',
    'Compliance', 'Capital', 'Equity', 'Debt', 'Liability', 
    'Contract', 'Regulation', 'Jurisdiction', 'Governance', 
    'Fraud', 'Penalty', 'Transaction', 'Asset', 'Treasury', 
    'Audit', 'Disclosure', 'Insolvency', 'Bankruptcy', 
    'Merger', 'Acquisition', 'Divestiture', 'Antitrust', 
    'Fiduciary', 'Interest', 'Dividend', 'Bond', 'Stock', 
    'Shareholder', 'Portfolio', 'Arbitration', 'Litigation', 
    'Reconciliation', 'Custodian', 'Brokerage', 'Underwriting', 
    'Hedge', 'Derivative', 'Swap', 'Option', 'Valuation', 
    'Prospectus', 'Collateral', 'Leverage', 'Liquidation', 
    'Monetary', 'Remittance', 'Escrow', 'Fiscal'
]
```

```python
keywords_lower = [keyword.lower() for keyword in keywords]
# Function to count keywords in content
def count_keywords(content):
    if isinstance(content, str):  # Check if content is a string
        content_lower = content.lower()
        return sum(1 for keyword in keywords_lower if keyword in content_lower)
    else:
        return 0  # Return 0 if content is not a string

# Apply the function to count keywords and calculate the points
tope_documentID_df_['keyword_count'] = tope_documentID_df_['Content'].apply(count_keywords)
tope_documentID_df_['keywords_point'] = tope_documentID_df_['keyword_count'] * 0.1

# Display the resulting DataFrame length
len(tope_documentID_df_)
```

```python
tope_documentID_df_['total_point'] = tope_documentID_df_['Points'] + tope_documentID_df_['keywords_point']
unique_regulators = tope_documentID_df_[tope_documentID_df_['total_point'] > 8]['RegulatorId'].unique()
len(unique_regulators)
```

```python
def predict_relevance(document, unique_regulators):
    
    """
    This system says a document is relevant if it comes from a regulator
    that is in the list of unique_regulators.
    """
    if document["RegulatorId"] in unique_regulators:
        return True
    return False
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import nltk
from nltk.corpus import stopwords
import re

# Download stopwords if you haven't already
nltk.download('stopwords')
```

```python
def clean_text(text):
    if isinstance(text, str):  # Only process if the text is a string
        # Remove HTML tags
        soup = BeautifulSoup(text, "html.parser")
        cleaned_text = soup.get_text()
        
        # Remove stop words
        stop_words = set(stopwords.words('english'))
        words = cleaned_text.split()
        filtered_text = ' '.join([word for word in words if word.lower() not in stop_words])
        
        # Optionally, you can remove any remaining punctuation or special characters
        filtered_text = re.sub(r'[^\w\s]', '', filtered_text)
        
        return filtered_text
    else:
        return ''  # Return an empty string if the text is NaN or not a string
```

```python
merged_data['Cleaned_Content'] = merged_data['Content'].apply(clean_text)
```
